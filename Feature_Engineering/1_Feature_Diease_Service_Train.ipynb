```python
df_disease["出院诊断病种名称_clean"] = df_disease["出院诊断病种名称"].str.replace("*"," ").\
                                        str.replace("-"," ").str.replace("，"," ").str.replace("、"," ").\
                                        str.replace("；"," ").str.replace("？"," ").str.replace("（"," ").\
                                        str.replace("）"," ").str.replace("："," ").\
                                        str.replace("[0-9,;-@+=?()a-zA-Z/:\.]"," ").\
                                        str.replace("(\[\W*\w*\])"," ").\
                                        str.strip()
```

```python
df_disease["出院诊断病种名称_clean"].fillna("未知",inplace=True)
```

```python
# 诊断病种的种类，包括nan取为未知
disease_set = []
for item in df_disease["出院诊断病种名称_clean"].str.split():
    disease_set.extend(item)
disease_set = list(set(disease_set))
```

```python
grp = df_disease.groupby(level=0)["出院诊断病种名称_clean"]
```

```python
len(disease_set)
```

```python
num_disease_type = len(disease_set)
def disease_2vector(disease_list):
    disease_vector = [0]*num_disease_type
    for disease in disease_list:
        disease_vector[disease_set.index(disease)] += 1
#    print disease_vector[1692]
    return disease_vector
```

```python
time_start = time.time()

df_disease_feat_list = []
index = []
#j = 1
for name, value in grp:
    disease_list = []
#    print name
    for item in value.str.split():
#        print item
        disease_list.extend(item)
    df_disease_feat_list.append(disease_2vector(disease_list))
    index.append(name)

#    j += 1
#    if j > 1000:
#        break

time_end=time.time()
print "Time: ", time_end-time_start
```

```python
df_disease_feat = pd.DataFrame(data=df_disease_feat_list,columns=disease_set,index=index)
df_disease_feat.index.name = "个人编码"
```

# 单价与数量
单价*数量；单价的统计量；数量的统计量

```python
 df_disease["单价"].mean()
```

```python
df_disease["单价"] = df_disease["单价"].fillna(df_disease["单价"].median())
```

```python
df_disease["数量"] = df_disease["数量"].fillna(df_disease["数量"].median())
```

```python
df_feat_price_stats = df_disease.groupby(level=0)["单价"].agg(['max','mean','median','std'])
df_feat_price_stats.columns = ["{}_{}".format("单价",col) for col in ['max','mean','median','std']]
```

```python
df_feat_price_stats
```

```python
df_feat_num_stats = df_disease.groupby(level=0)["数量"].agg(['max','mean','median','std'])
df_feat_num_stats.columns = ["{}_{}".format("数量",col) for col in ['max','mean','median','std']]
```

```python
df_feat_num_stats
```

```python
df_disease["total_sevice_cost"] = df_disease["数量"]*df_disease["单价"]
```

```python
df_feat_sevicecost_stats = df_disease.groupby(level=0)["total_sevice_cost"].agg(['max','mean','sum','min','median','std'])
df_feat_sevicecost_stats.columns = ["{}_{}".format("sevice_cost",col) for col in ['max','mean','sum','min','median','std']]
```

```python
df_feat_sevicecost_stats
```

```python
feat_service_stats = pd.concat([df_feat_price_stats,df_feat_num_stats,df_feat_sevicecost_stats],axis=1)
```

```python
feat_service_stats.to_csv("feat_service_stats_train.csv")
```
