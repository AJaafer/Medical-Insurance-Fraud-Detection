# 各个条目统计

```python
item_list = []
for line in open("feature_item.txt","r"):
    item_list.append(line.strip())
for item in item_list:
    print "sum({}) as sum_{}, avg({}) as avg_{}, median({}) as mid_{},stddev({}) as std_{}, max({}) as max_{},".format(item,item,item,item,item,item,item,item,item,item)
```

```python
disease_list = []
for line in open("feat_disease_selected.csv","r"):
    disease_list.append(line.split(",")[1].strip())
for i,disease in enumerate(disease_list[1:]):
    print 'sum(case when disease regexp "{}" then 1 else 0 end) as disease_type_{},'.format(disease,i+1)
```

# 重要三目

```python
service_list = []
for line in open("feat_service_selected.csv","r"):
    service_list.append(line.split(",")[1].strip())
for i,service in enumerate(service_list[1:]):
    print 'sum(case when (sm_proj_name regexp "{}" or sm_hos_proj_name regexp "{}") then 1 else 0 end) as service_type_{},'.format(service,service,i+1)
```
