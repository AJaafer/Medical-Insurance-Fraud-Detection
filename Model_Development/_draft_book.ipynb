```python
a = [1,2,6,1]; b = [1,2,6]
def a_vs_b(a,b):
    i = 1
    while(a[i-1]==b[i-1] and i<len(a) and i<len(b)):
        i += 1
    if (i==len(a) and i==len(b)):
        print "a = b"
    else:
        if i==len(a):
            print "a < b"
            return
        if i==len(b):
            print "a > b"
            return
        if a[i] > b[i]:
            print "a > b"
        else:
            print "a < b"
    return

a_vs_b(a,b)           
        
```
