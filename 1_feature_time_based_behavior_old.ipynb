### *去医院的时间间隔或频率统计
最小，最大，平均，连续天数  
rate = freq/timeperiod  

能否计算出去医院间隔数的频率，此时设为间隔为0-365天，统计各个间隔下，每人出现的频率，共365列，一个间隔一个特征  
太稀疏，是否需要分箱统计？

```python
# 每次去医院的最大时间间隔max
def every_max_period(s):
    return s.sort_values().diff(1).astype('timedelta64[D]').max()
# 每次去医院的最小时间间隔min
def every_min_period(s):
    return s.sort_values().diff(1).astype('timedelta64[D]').min()
# 每次去医院的平均时间间隔ave
def every_ave_period(s):
    return s.sort_values().diff(1).astype('timedelta64[D]').mean()
# 每次去医院的时间间隔4分位数
def every_q1_period(s):
    return s.sort_values().diff(1).astype('timedelta64[D]').quantile(0.25)
# 每次去医院的时间间隔4分位数
def every_q2_period(s):
    return s.sort_values().diff(1).astype('timedelta64[D]').median()
# 每次去医院的时间间隔4分位数
def every_q3_period(s):
    return s.sort_values().diff(1).astype('timedelta64[D]').quantile(0.75)

# 去医院的总的时间跨度
def total_period(s):
    return (s.max()-s.min()).days
# 去医院的总的频率
def total_rate(s):
    return (s.max()-s.min()).days/float(len(s))
```
